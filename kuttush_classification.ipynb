```python
from keras.applications.inception_v3 import InceptionV3, preprocess_input, decode_predictions
from keras.preprocessing import image
import requests
import numpy as np
model = InceptionV3(include_top=True, weights='imagenet', input_tensor=None, input_shape=None)
file_name = 'Kuttush.png'
```

```python
img = image.load_img(file_name, target_size=(299, 299))
img
```

```python
ary = image.img_to_array(img)
ary = np.expand_dims(ary, axis=0)
```

```python
preds = model.predict(preprocess_input(ary))
results = decode_predictions(preds, top=10)[0]
for result in results:
    print(result)
```
